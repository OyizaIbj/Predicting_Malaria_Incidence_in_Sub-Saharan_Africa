```python
# Import required libraries
import pandas as pd

# Load the Datasets
geo_df = pd.read_csv('Geospatial_Information.csv')
malaria_df = pd.read_csv('Malaria_Data.csv')
climate_df = pd.read_csv('Climatic_Data.csv')
socio_df = pd.read_csv('Socioeconomic_Data.csv')

print("Datasets Loaded Successfully")
```

```python
# Merge the Datasets on Country, Year, and Month
merged_df = pd.merge(geo_df, malaria_df, on=['Country', 'Year', 'Month'], how='inner')
merged_df = pd.merge(merged_df, climate_df, on=['Country', 'Year', 'Month'], how='inner')
merged_df = pd.merge(merged_df, socio_df, on=['Country', 'Year', 'Month'], how='inner')

print("Datasets Merged Successfully")
```

#### I used **inner join (how='inner')**, to keep only the rows that exist in all datasets.

```python
# Quick Data Check
print("Merged Dataset Shape:", merged_df.shape)
print("\nMerged Dataset Columns:\n", merged_df.columns.tolist())
print("\nSample Data:\n") 
merged_df.head()
```

```python
# Save Merged Dataset
merged_df.to_csv('Malaria_Project_Data.csv', index=False)
print("Merged dataset saved as 'Malaria_Project_Data.csv'")
```
